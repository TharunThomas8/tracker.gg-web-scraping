```python
from selenium import webdriver
import requests
from bs4 import BeautifulSoup
import time
from datetime import datetime
from selenium.common.exceptions import NoSuchElementException
from datetime import date
from selenium.webdriver.common.by import By
import pandas as pd
```

```python
url = 'https://tracker.gg/valorant/lfg'
test = 'https://www.google.com/'
browser = webdriver.Chrome(executable_path='D:/Study/TCD/ML/grp/chromedriver')
table_xpath = '/html/body/div/div[2]/div[2]/div/main/div[3]/div[3]/div[2]/div/div[2]/table'

browser.get(url)
time.sleep(5)


matchs = browser.find_elements(By.TAG_NAME, 'tr')

rank = []
mode = []
win_per = []
kd = []
dmg_per_rnd = []



for m in matchs:

    # username.append(m.find_elements(By.XPATH, './/td[1]')[0].text)
    for u in m.find_elements(By.XPATH, './/td[2]'):
        # print(u.text+"-----")
        rank.append(u.text)
    
    for u in m.find_elements(By.XPATH, './/td[3]'):
        # print(u.text+"-----")
        mode.append(u.text)
    
    for u in m.find_elements(By.XPATH, './/td[4]'):
        # print(u.text+"-----")
        win_per.append(u.text)
    
    for u in m.find_elements(By.XPATH, './/td[5]'):
        # print(u.text+"-----")
        kd.append(u.text)
    
    for u in m.find_elements(By.XPATH, './/td[6]'):
        # print(u.text+"-----")
        dmg_per_rnd.append(u.text)


player_tag_xpath = "/html/body/div/div[2]/div[2]/div/main/div[3]/div[3]/div[2]/div/div[2]/table/tbody/tr/td[1]/div/span"
player_tag = browser.find_elements(By.XPATH, player_tag_xpath)
# matchs = browser.find_elements(By.XPATH, './/td[1]')
# f = 0
# i = 1
id = []

for i in player_tag:
    # ids.append(i.text)
    id.append((str(i.text).replace("\n", "")))
    # print(str(i.text))

region = []
cnt_of_child = 0
pr = '/html/body/div/div[2]/div[2]/div/main/div[3]/div[3]/div[2]/div/div[2]/table/tbody/tr/td[1]/div/div'
parnt = browser.find_elements(By.XPATH,pr)
# print(matchs)
for m in parnt:
    # for u in m.find_elements(By.XPATH, './/img[1]'):
    #     print((m.get_attribute("src")))
    cnt_of_child = 0
    for u in m.find_elements(By.XPATH, './/child::img'):
        # print((m.get_attribute("class")))
        # print("true")
        cnt_of_child = 1+cnt_of_child
    
    if(cnt_of_child==2):
        a_c = m.find_elements(By.XPATH,("./*"))
        f = 0
        for c in a_c:
            if(f==1):
                region.append(c.get_attribute("title"))
            f = f + 1
    else:
        region.append("None")
            # print("here")
    # print("-------")

browser.quit()

# print(id)
print(len(id))
print(len(rank))
print(len(mode))
print(len(win_per))
print(len(kd))
print(len(dmg_per_rnd))
print(len(region))

```

```python
acc_head_per = []
acc_head_hits = []
acc_body_per = []
acc_body_hits = []
acc_leg_per = []
acc_leg_hits = []

matchs = []
headshot_per = []
wins = []
kills = []
kill_per_rnd = []

top_agent_name = []
top_agent_time = []
top_agent_matchs = []
top_agent_win_per = []
top_agent_kd = []
top_agent_adr = []

num = 0
for i in id:
    num= num+1
    id_values = i.split("#")
    # print(id_values[0],id_values[1])
    url = 'https://tracker.gg/valorant/profile/riot/'+id_values[0]+'%23'+id_values[1]+'/overview?playlist=competitive&season=all'

    browser = webdriver.Chrome(executable_path='D:/Study/TCD/ML/grp/chromedriver')

    browser.get(url)
    # browser.maximize_window()
    time.sleep(3)
    browser.minimize_window()
    

    try:
        parnt = browser.find_element(By.XPATH,"//*[@class= 'accuracy__stats']/tbody/tr[1]")
        sep = (parnt.text).split(" ")
        # print(sep[1],sep[2]+"--------")
        acc_head_per.append((sep[1]).replace('%',''))
        acc_head_hits.append((sep[2]))

        parnt = browser.find_element(By.XPATH,"//*[@class= 'accuracy__stats']/tbody/tr[2]")
        sep = (parnt.text).split(" ")
        # print(sep[1],sep[2]+"--------")
        acc_body_per.append((sep[1]).replace('%',''))
        acc_body_hits.append((sep[2]))

        parnt = browser.find_element(By.XPATH,"//*[@class= 'accuracy__stats']/tbody/tr[3]")
        sep = (parnt.text).split(" ")
        # print(sep[1],sep[2]+"--------")
        acc_leg_per.append((sep[1]).replace('%',''))
        acc_leg_hits.append((sep[2]))

        parnt = browser.find_element(By.XPATH,"//*[@class= 'segment-stats area-main-stats card bordered header-bordered responsive']/div/div[1]/div[1]/span[2]")
        sep = (parnt.text).split(" ")
        # print(sep[0]+"--------")
        matchs.append((sep[0]))


        parnt = browser.find_element(By.XPATH,"//*[@class= 'giant-stats']/div[3]/div/div/span[2]")
        sep = (parnt.text).split("%")
        # print(sep[0]+"--------")
        headshot_per.append(sep[0])

        parnt = browser.find_elements(By.XPATH,"//*[@class= 'main']/div/div/div/span[2]")
        wins.append(parnt[0].text)
        kills.append(parnt[1].text)
        kill_per_rnd.append(parnt[6].text)
        
        parnt = browser.find_elements(By.XPATH,"//*[@class= 'top-agents area-top-agents']/div/div/div/div[2]/div/div[1]/div")
        # print(parnt[0].text)
        top_agent_name.append(parnt[0].text)
        
        sep = (parnt[1].text).split(" ")
        top_agent_time.append(sep[0])
        # print(sep[0])
        top_agent_matchs.append(parnt[2].text)
        # print(parnt[2].text)
        sep = (parnt[3].text).split("%")
        # print(sep[0])
        top_agent_win_per.append(sep[0])
        # print(parnt[4].text)
        top_agent_kd.append(parnt[4].text)
        # print(parnt[5].text)
        top_agent_adr.append(parnt[5].text)

        # print("-----------")
        print(" Done " + i)
        print(num)

    except NoSuchElementException:
        acc_head_per.append('0')
        acc_head_hits.append('0')
        acc_body_per.append('0')
        acc_body_hits.append('0')
        acc_leg_per.append('0')
        acc_leg_hits.append('0')
        matchs.append("0")
        headshot_per.append('0')
        wins.append('0')
        kills.append('0')
        kill_per_rnd.append('0')
        top_agent_name.append('0')
        top_agent_time.append('0')
        top_agent_matchs.append('0')
        top_agent_win_per.append('0')
        top_agent_kd.append('0')
        top_agent_adr.append('0')

        print("Failed "+i)
        

    browser.quit()
    
    

# print((acc_head_per),(acc_head_hits),acc_body_per,acc_body_hits,acc_leg_per,acc_leg_hits)
```
